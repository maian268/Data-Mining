# 01 - Khám phá dữ liệu (EDA - Insight)

Phân tích cấu trúc, chất lượng, thống kê mô tả healthcare stroke dataset.

## 1. Đọc dữ liệu và import thư viện

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', None)

data_path = '../data/raw/healthcare-dataset-stroke-data.csv'
df = pd.read_csv(data_path)
print(f'Shape: {df.shape}')
df.head()
```

## 2. Kiểm tra thông tin tổng quan, giá trị thiếu, kiểu dữ liệu

```python
# Thông tin tổng quan và giá trị thiếu
df.info()
missing = df.isna().sum().to_frame('missing')
missing['pct'] = missing['missing'] / len(df) * 100
print(missing[missing['missing']>0].sort_values('missing', ascending=False))
```

```python
# 1. Xử lý thiếu cho bmi + tạo cột cờ
df['bmi_missing'] = df['bmi'].isna().astype(int)
df['bmi'] = df['bmi'].fillna(df['bmi'].median())

# 2. Chuyển object → category
cat_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for c in cat_cols:
    df[c] = df[c].astype('category')

# 3. Loại bỏ cột id (không mang thông tin dự báo)
df = df.drop(columns=['id'])

# Kiểm tra lại
df.info()
df.isna().sum()
```

## 3. Thống kê mô tả, kiểm tra phân phối, outlier

### Nhận xét nhanh (thống kê & phân phối)
-  Biến số trung bình/median đáng chú ý: 
    + Age trung bình 43 (median 45), phân bố khá đều nhưng có trẻ sơ sinh (min 0.08); 
    + Bmi trung bình 28.9 (median 28.1) → dân số hơi thừa cân, có outlier cực đại 97.6;
    + Avg_glucose_level lệch phải rất mạnh (trung bình 106, median 91, max 271) → nhiều bệnh nhân đái tháo đường nặng
- Biến phân loại có tần suất lệch nhiều:
    + Nữ chiếm 59%, nam 41%, có 1 mẫu “Other”; 
    + Work_type “Private” chiếm 57%, “children” và “Never_worked” rất ít; 
    + Smoking_status có tới 30% “Unknown” → thực chất là giá trị thiếu ngầm, cần xử lý như missing.

```python
# Thống kê mô tả
numeric_desc = df.describe(include=[np.number]).T
display(numeric_desc)

# Categorical summary
for col in df.select_dtypes(include=['category']).columns:
    print(f"\n{col}:")
    print(df[col].value_counts(dropna=False))

# Vẽ histogram cho các biến số
for col in df.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(5,2))
    plt.hist(df[col].dropna(), bins=30, edgecolor='black')
    plt.title(f'Histogram: {col}')
    plt.show()

# Boxplot kiểm tra outlier
for col in df.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(5,2))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot: {col}')
    plt.show()
```

### Nhận xét nhanh (cân bằng lớp)
Tỷ lệ lớp 
 Stroke=1 chỉ chiếm 4.87% (249/5110), Stroke=0 chiếm 95.13% 

→ tỷ lệ mất cân bằng cực kỳ nghiêm trọng (1:20).


Nếu không xử lý, mô hình sẽ đạt accuracy 95% chỉ bằng cách đoán tất cả là 0.

→ Bắt buộc phải dùng một trong các cách sau: Class_weight='balanced' , SMOTE/ADASYN, hoặc các thuật toán chuyên cho imbalance (Balanced RF, EasyEnsemble…).

## 4. Kiểm tra cân bằng lớp (Class Balance)
- Kiểm tra phân phối biến mục tiêu (target: 'stroke')
- Vẽ biểu đồ bar cho biến mục tiêu

### Nhận xét nhanh (tương quan)
- Cặp biến tương quan cao (>|0.7|): 

Không có cặp biến số nào đạt ngưỡng đa cộng tuyến nghiêm trọng (tất cả |corr| < 0.5). Cặp cao nhất chỉ là age ↔ ever_married (~0.68), hoàn toàn chấp nhận được → không cần loại bỏ biến nào.

- Biến liên quan đến target: 

age: corr = 0.25 (mạnh nhất, rủi ro tăng rõ theo tuổi)
hypertension: 0.13
heart_disease: 0.13
avg_glucose_level: 0.13
bmi: chỉ 0.04 (rất yếu)
→ Top 4 yếu tố nguy cơ rõ rệt: tuổi cao, tăng huyết áp, bệnh tim, đường huyết cao.
BMI tuy yếu về tương quan tuyến tính nhưng vẫn có thể quan trọng trong mô hình phi tuyến (Random Forest, XGBoost…).

- Bonus nhận xét:

Các biến phân loại (sau khi encode) cũng cho thấy sự khác biệt rõ:
formerly smoked và smokes có tỷ lệ stroke cao hơn never smoked
Self-employed và Private có nguy cơ cao hơn Govt_job
→ Sau khi one-hot/label encode, các biến này chắc chắn sẽ được mô hình tree-based chọn làm feature quan trọng.

**Tổng kết**: Không lo đa cộng tuyến, nhưng stroke phụ thuộc mạnh vào age và các bệnh nền → tập trung khai thác tốt các biến này + xử lý imbalance là đủ để có mô hình hiệu quả.


```python
# Kiểm tra phân phối biến mục tiêu (stroke)
class_counts = df['stroke'].value_counts()
print(class_counts)

plt.figure(figsize=(4,3))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Class Balance: stroke')
plt.xlabel('Stroke')
plt.ylabel('Count')
plt.show()
```

## 5. Phân tích tương quan (Correlation Analysis)
- Tính ma trận tương quan cho các biến số
- Vẽ heatmap để trực quan hóa mối quan hệ giữa các biến

```python
# Ma trận tương quan và heatmap
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

- Ma trận tương quan cho thấy age có tương quan dương mạnh nhất với stroke (khoảng 0.25), theo sau là hypertension (0.13) và heart_disease (0.13), cho thấy tuổi cao và các bệnh mãn tính là yếu tố rủi ro chính.

- Các biến khác như avg_glucose_level và bmi có tương quan yếu với stroke (dưới 0.1), nhưng có thể có tương tác phi tuyến. Không có multicollinearity mạnh giữa các biến độc lập (tất cả corr < 0.5), nên phù hợp cho mô hình tuyến tính. Biến bmi_missing (cờ missing) có corr thấp, cho thấy missing ở bmi không ảnh hưởng lớn đến stroke.

## 6. Tổng kết EDA & hành động tiếp theo


- Tổng kết

    - Dataset có 5110 mẫu, 11 biến (sau drop id), với missing chỉ ở bmi (3.93%, đã xử lý bằng median).

    -Phân phối: Tuổi và BMI lệch phải (outlier ở glucose/BMI), categorical lệch về nữ, married, private work, và smoking unknown cao (có thể là dữ liệu thiếu ngầm cần kiểm tra thêm).

    - Stroke imbalance nghiêm trọng (4.87% positive), cần xử lý để tránh bias mô hình.

    - Tương quan: Age, hypertension, heart_disease là các yếu tố chính liên quan đến stroke; không có corr cao giữa biến độc lập.

    - Phân phối tổng quát: Mẫu đa dạng độ tuổi (0-82), nhưng tập trung vào người lớn (median 45), và rủi ro stroke tăng theo tuổi/bệnh nền.

- Liệt kê hành động:
    
    - Xử lý missing: Đã fill bmi bằng median và tạo cờ bmi_missing; kiểm tra thêm unknown ở smoking_status (có thể coi là missing và impute mode "never smoked").

    - Outlier: Capping glucose/BMI tại percentile 99 (ví dụ: glucose >250 → 250) hoặc transform log để giảm lệch.

    - Cân bằng lớp: Sử dụng class_weight trong mô hình, hoặc oversampling SMOTE cho lớp stroke=1.


    - Encode/cast dtype: One-hot encode categorical (gender, work_type, etc.), giữ numerical như float/int; chia train-test với stratify=y để giữ tỷ lệ stroke.

    - Tiếp theo: Feature engineering (ví dụ: bin age thành nhóm <18, 18-60, >60), thử mô hình baseline (Logistic Regression/Random Forest) và đánh giá bằng F1/Recall do imbalance.
