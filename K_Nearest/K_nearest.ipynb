# 2. CHIA TRAIN / TEST 80%/20%

```python
split = int(0.8 * len(all_data))
train_data = all_data[:split]
test_data  = all_data[split:]

# Tạo dict data để predict() dùng
data = {}
for f, label in train_data:
    data.setdefault(label, []).append(f)
```

# 3. HÀM DỰ ĐOÁN KNN + CONFIDENCE

```python
def predict(input_feature, k):
    input_feature = list(map(float, input_feature))
    distances = []

    for group in data:
        for train_feature in data[group]:
            train_feature = list(map(float, train_feature))
            euclidean = np.linalg.norm(np.array(input_feature) - np.array(train_feature))
            distances.append([euclidean, group])

    nearest = sorted(distances)[:k]
    votes = [d[1] for d in nearest]

    counter = Counter(votes)
    prediction, count = counter.most_common(1)[0]

    confidence = count / k * 100  # phần trăm
    return prediction, confidence
```

# 4. TÍNH ACCURACY

```python
correct = 0
for f, true_label in test_data:
    pred = predict(f, k=5)
    if pred == true_label:
        correct += 1

accuracy = correct / len(test_data) * 100
print("Accuracy:", accuracy, "%")
```
